```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
from pandas import DataFrame, read_csv, read_sql_query
from collections import Counter
import pandas as pd #this is how I usually import pandas
import sys #only needed to determine Python version number
import numpy as np
db = sqlite3.connect('tmp/datajobs.db')
cur = db.cursor()

df = pd.read_sql_query("SELECT title, description FROM job ORDER BY published DESC;", db)
```

```python
print df.head()
counter = Counter(" ".join(df["description"]).split()).most_common(150)
words = np.array([pair[0] for pair in counter])
freqs = np.array([pair[1] for pair in counter])
fig = plt.figure(figsize=(20, 10))


print len(freqs)
# Plot histogram using matplotlib bar()
plt.xticks(range(len(freqs)), words, rotation=90)
plt.bar(range(len(freqs)), freqs)
plt.show()
```
